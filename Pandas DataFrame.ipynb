```python
data.fillna(75, inplace=True)   # 비 열이 있으면 그 열에 데이터를 75로 채워준다
data['국어'].fillna(73, inplace=True)   # 국어에 0값이 있으면 데이터를 73으로 채워준다
data
```

```python
avg = data['국어'].mean()    # 국어과목 평균 구하기
avg
```

```python
median = data['국어'].median()   # 오름차순 정렬후 가운데 중앙값을 가져오기
median
```

```python
most = data['국어'].mode()[0]    # 국어 점수중에서 가장 많이 나타나는 점수 한개를 고름
most
```

```python
data['국어'].fillna(most, inplace=True)
```

### Raplacing Value

```python
data.loc[0, '국어'] = 90     # 0번째 인덱스의 '국어' 점수를 90으로 수정
data
```

```python
for i in data.index:
    if data.loc[i, '국어'] > 100:   # 모든 열에서 '국어' 점수가 100보다 크면
        data.loc[i, '국어'] = 100   # 최고 100점으로 맞춤
        
data
```

```python
data.duplicated()    # 중복값이 있는지 검사
```

```python
data.drop_duplicates(inplace=True)    # 중복값이 있으면 제거
```

```python
data.corr(numeric_only=True)
```

```python
# 총점 계산
data['총점'] = data[['국어', '영어', '수학', '과학']].sum(axis=1)

# 평균 계산
data['평균'] = data[['국어', '영어', '수학', '과학']].mean(axis=1)

data
```

```python
data['학년/반'] = ['' for _ in range(30)]     # 학년/반 이라는 컬럼 삽입하고 값을 ''(빈문자열)로 초기화하기
data
```

```python
data.rename(columns={'학년/반':'반'}, inplace=True)    # 컬럼의 이름을 바꾸기
data
```

```python
for i in range(30):               # 모든 열을 순회하면서 반 컬럼의 값을 'A', 'B', 'C'로 삽입하기
    if i < 10:
        data.loc[i, '반'] = 'A'
    elif i>=10 and i<20:
        data.loc[i, '반'] = 'B'
    else:
        data.loc[i, '반'] = 'C'
data
```

```python
data.columns
```

```python
data = data[['no', 'name', '반','국어', '영어', '수학', '과학', '총점', '평균']]
data
```

```python
grouped = data.groupby('반')[['국어','총점','평균']].mean()
grouped
```
