### Pandas로 데이터셋 핸들링하기

```python
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
org_data = {
    "이름":['유재석','지석진','김종국','송지효','하하','양세찬','전소민','강호동','이수근','이상민'],
    "소속":['런닝맨','런닝맨','런닝맨','런닝맨','런닝맨','런닝맨','런닝맨','아형','아형','아형'],
    "국어":[random.randint(40,100) for _ in range(10)],
    "영어":[random.randint(40,100) for _ in range(10)],
    "수학":[random.randint(40,100) for _ in range(10)],
    "과학":[random.randint(40,100) for _ in range(10)],
}

data = pd.DataFrame(org_data)
```

```python
data
```

```python
names = data['이름']       # 얻고자하는 필드가 하나의 컬럼일경우 []에 한개의 컬럼을 직접 입력할 수 있다.
names
```

```python
data['이름', '국어']    # 얻고자하는 데이터를 리스트에 담아서 전달해야 되므로 이렇게 하면 안되고 
                      # [[]] 형식을 사용해야 된다.
```

```python
data[['이름','국어']]     # 선택한 컬럼만 가져오기
```

```python
data.loc[1:3]          # 열의 슬라이싱
```

```python
data.loc[0,'영어']=60       # 한개의 데이터 접근 및 수정
data.loc[0,'영어']
```

```python
# 총점 계산
data['총점'] = data[['국어', '영어', '수학', '과학']].sum(axis=1)     # 새로운 컬럼 삽입하기

# 평균 계산
data['평균'] = data[['국어', '영어', '수학', '과학']].mean(axis=1)

data
```

```python
# 총점으로 정렬하기, 기본은 오름차순
data.sort_values(by='총점')
```

```python
# 총점으로 내림 차순 정렬하기
data.sort_values(by='총점', ascending=False)
```

```python
def grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'
    
data['평가'] = data['평균'].apply(grade)
data
```

```python
# 상위 3명만 뽑기
top3 = data['총점'].quantile(0.7)
top3_data = data[data['총점']>=top3]
top3_data.sort_values(by='총점', ascending=False)
```

```python
# 필터링
runningman = data[data['소속']=='런닝맨']
runningman
```

```python
# 필터링
bottoms = data[data['평균']<70]
bottoms
```

```python
# 그룹핑 : 필터링과는 다른 개념으로 그룹을 만들고 그룹으로 의미있는 데이터를 생성하여 보여주는데 목적이 있음

grouped = data.groupby('소속')
grouped.mean(numeric_only=True).round(2)
```

### Matplotlib을 이용한 그래프 출력

```python
import matplotlib as mpl
mpl.rcParams['font.family'] = 'AppleGothic'
mpl.rcParams['font.size'] = 10
mpl.rcParams['axes.unicode_minus'] = False    # 마이너스 부호를 깨지지 않게 표시함

x = data['이름']
y = data['평균']
plt.plot(x,y)
```

```python
plt.bar(x,y)
```

```python
x = data['이름']
y1 = data['국어']
y2 = data['영어']
y3 = data['수학']
y4 = data['과학']
plt.bar(x,y1)
plt.bar(x,y2)
plt.bar(x,y3)
plt.bar(x,y4)
```

```python
x = data['이름']
y1 = data['국어']
y2 = data['영어']
y3 = data['수학']
y4 = data['과학']
plt.bar(x,y1)
plt.bar(x,y2, bottom=y1)
plt.bar(x,y3, bottom=y1+y2)
plt.bar(x,y4, bottom=y1+y2+y3)
```

```python
# 누적 막대그래프
y1 = data['국어']
y2 = data['영어']
y3 = data['수학']
y4 = data['과학']
plt.bar(data['이름'],data['국어'], label='국어')
plt.bar(data['이름'],data['영어'], bottom=data['국어'], label='영어')
plt.bar(data['이름'],data['수학'], bottom=data['국어']+data['영어'], label='수학')
plt.bar(data['이름'],data['과학'], bottom=data['국어']+data['영어']+data['수학'], label='과학')
#plt.legend(['국어','영어','수학','과학'])     # 범례를 보여줌
plt.legend()
```

```python
# pandas를 이용한 다중 막대그래프 그리기
scores = data.set_index('이름')
scores.plot(kind='bar')
plt.legend()
```

```python
# 총점을 빼고 다중 막대그래프 그리기
without_total = data.drop('총점', axis=1)
scores = without_total.set_index('이름')
scores.plot(kind='bar')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # 좌상단을 기준으로 x축으로 5%만큼 이동 y축은 그대로인 위치
```

```python
# 산점도 그래프 그리기 
plt.scatter(data['국어'], data['영어'])
plt.xlabel('국어')
plt.ylabel('영어')
```

```python
# 다중막대그래프 Customizing

# 그래프의 기본 가로 세로 사이즈 알아보기
plt.rcParams['figure.figsize']  # inch단위로 반환
```

```python
index = data.index
plt.bar(index-0.4, data['국어'], label='국어', width=0.2)
plt.bar(index-0.2, data['영어'], label='영어', width=0.2)
plt.bar(index, data['수학'], label='수학', width=0.2)
plt.bar(index+0.2, data['과학'], label='과학', width=0.2)
```

```python
index = data.index
plt.bar(index-0.4, data['국어'], label='국어', width=0.2)
plt.bar(index-0.2, data['영어'], label='영어', width=0.2)
plt.bar(index, data['수학'], label='수학', width=0.2)
plt.bar(index+0.2, data['과학'], label='과학', width=0.2)
plt.xticks(index, data['이름'], rotation=90)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # 좌상단을 기준으로 x축으로 5%만큼 이동 y축은 그대로인 위치
plt.show()
```
