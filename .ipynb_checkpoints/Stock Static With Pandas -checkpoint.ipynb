```python
import pandas as pd
```

```python
data = pd.read_excel('/Users/eunbumkim/Documents/Practice/python/pandas/stock.xlsm')
data.head(10)
```

```python
data.tail()
```

```python
pivot_data = data.pivot_table(index='종목', columns='날짜', values='전일비')
pivot_data
```

```python
grouped = data.groupby('종목')[['전일비','거래량']].sum(numeric_only=True)
grouped
```

```python
fmt_data = data.applymap(lambda x: '{:,.0f}'.format(x) if isinstance(x, (int, float)) else x)
fmt_data.head()
```

```python
fmt_data.describe()
```

```python
fmt_data.head()
```

```python
fmt_data[['종목','종가']].head()
```

```python
group_data = data.groupby('종목')[['거래량','전일비']].sum()
fmt_data = group_data.applymap(lambda x: '{:,}'.format(x) if isinstance(x, (int, float)) else x) 
fmt_data
```

```python
group_data = data.groupby('종목')[['거래량','전일비']].mean()
fmt_data = group_data.applymap(lambda x: '{:,.00f}'.format(x) if isinstance(x, (int, float)) else x) 
fmt_data
```

```python
import matplotlib.pyplot as plt
```

```python
x = pivot_data.index          # x축은 날짜로 함
y = pivot_data[['삼성전자']]    # y축은 삼성전자 컬럼이 저장된 값
plt.plot(x,y)
```

```python
x = pivot_data.index
y = pivot_data[['삼성전자','SK하이닉스','LG에너지솔루션']]
plt.plot(x,y)
```

```python
x = filter_data['날짜']
y = filter_data['종가']
plt.plot(x,y)
```

```python
x = data[data['종목']=='삼성전자']['날짜']
y = data[data['종목']=='SK하이닉스']['종가']
plt.plot(x,y)
```

```python
plt.title('SK하이닉스 종가 변동 추이')
plt.plot(x,y)
```

```python
import matplotlib as mpl
mpl.rcParams['font.family'] = 'AppleGothic'
mpl.rcParams['font.size'] = 10
mpl.rcParams['axes.unicode_minus'] = False    # 마이너스 부호를 깨지지 않게 표시함
plt.title('SK하이닉스 종가 변동 추이')
plt.plot(x,y)
```

```python
import matplotlib.font_manager as fm
fm.fontManager.ttflist                      # 사용가능한 폰트 출력
[f.name for f in fm.fontManager.ttflist]    # 사용가능한 폰트중에서 이름 부분만 출력
```

```python
plt.plot(x,y)
plt.title('SK하이닉스 종가 변동 추이', fontdict={'family':'AppleGothic', 'size':14})
plt.xlabel('날짜', color='r', loc='right')     # 컬러값은 16진수나 'red'로 변경 가능
plt.ylabel('종가', color='b', loc='top')       # loc는 레이블의 위치
```
