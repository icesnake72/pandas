```python
import pandas as pd
```

```python
data = pd.read_excel('/Users/eunbumkim/Documents/Practice/python/pandas/stock.xlsm')
data.head(10)
```

```python
data.tail()
```

```python
pivot_data = data.pivot_table(index='종목', columns='날짜', values='전일비')
pivot_data
```

```python
grouped = data.groupby('종목')[['전일비','거래량']].sum(numeric_only=True)
grouped
```

```python
fmt_data = data.applymap(lambda x: '{:,.0f}'.format(x) if isinstance(x, (int, float)) else x)
fmt_data.head()
```

```python
fmt_data.describe()
```

```python
fmt_data.head()
```

```python
fmt_data[['종목','종가']].head()
```

```python
group_data = data.groupby('종목')[['거래량','전일비']].sum()
fmt_data = group_data.applymap(lambda x: '{:,}'.format(x) if isinstance(x, (int, float)) else x) 
fmt_data
```

```python
group_data = data.groupby('종목')[['거래량','전일비']].mean()
fmt_data = group_data.applymap(lambda x: '{:,.00f}'.format(x) if isinstance(x, (int, float)) else x) 
fmt_data
```

```python
filter_data = data[data['종목']=='삼성전자']
filter_data
```

```python
filter_data['전일비'].sum()
```

```python
filter_data['전일비'].mean()
```

```python
filter_data['종가'].std()
```
